### Cohesion test

```python
results_cohesion = pd.read_json("../results/experiment_W2V_several_runs_WEAT_2/WEAT/cohesion_test_3.json")
results_cohesion.head(3)
```

```python
# explicit attrib and target set names
results_cohesion = results_cohesion.merge(results_cohesion.set_names.apply(lambda sn: pd.Series(sn)), 
                                          left_index=True, right_index=True)
results_cohesion = results_cohesion.drop(columns=['test_name'])
```

```python
def aggregate_results_cohesion(cohesion_attributes): 
        
    signifs = []
    cohesion_attributes = np.array(cohesion_attributes).T
    for ps in cohesion_attributes:
        
        ps = np.array(ps)
        
        if (ps<=0.05).all():
            ps = "**"
        elif (ps<=0.10).all():
            ps = "*"
        else:
            ps = ""
        signifs.append(ps)
        
    return signifs[-1]
        
```

```python
results_cohesion_groups = results_cohesion.groupby(['A', 'B', 'X', 'Y', 'weat_test_num', 'corpus'])\
                                            .apply(lambda rows: 
                                          pd.Series({'cohesion_attributes':aggregate_results_cohesion(rows.p_cohesion_attributes.tolist()), 
                                                     'cohesion_targets':aggregate_results_cohesion(rows.p_cohesion_targets.tolist())}) )
                                                   
#results_cohesion_groups = results_cohesion_groups.reset_index()   
results_cohesion_groups.sort_index(level=['weat_test_num'])
                                                   
```

### WEAT

```python
results_weat = pd.read_json("../results/experiment_W2V_several_runs_WEAT_2/WEAT/weat_test_3.json")
results_weat.head(3)
```

```python
# explicit attrib and target set names
results_weat = results_weat.merge(results_weat.set_names.apply(lambda sn: pd.Series(sn)), 
                                          left_index=True, right_index=True)
results_weat = results_weat.drop(columns=['test_name'])
```

```python
def aggregate_results_weat(weat): 
    
    effect_size = weat[0]['effect_size']
    ps = np.array([w['p'] for w in weat])
    
    if (ps<=0.05).all():
        ps = "**"
    elif (ps<=0.10).all():
        ps = "*"
    else:
        ps = ""
        
    return {'effect_size':effect_size, 'p':ps}
    
    
```

```python
results_weat_groups = results_weat.groupby(['A', 'B', 'X', 'Y', "weat_test_num", "corpus"])\
                                .apply(lambda rows: 
                                      pd.Series(aggregate_results_weat(rows.sort_values('iter').weat_test.tolist())) )
#results_weat_groups = results_weat_groups.reset_index()
results_weat_groups.sort_index(level=['weat_test_num'])
```

#### SC-WEAT

```python
def aggregate_results_scweat(weat): 
    
    res = {}
    for i in ['X', 'Y']:
    
        score = weat[0][i]['score']
        ps = np.array([w[i]['p'] for w in weat])

        if (ps<=0.05).all():
            ps = "**"
        elif (ps<=0.10).all():
            ps = "*"
        else:
            ps = ""
            
        
        res[f'score_{i}'] = score
        res[f'p_{i}'] = ps
    return res
    
    
```

```python
results_scweat_groups = results_weat.groupby(['A', 'B', 'X', 'Y', "weat_test_num", "corpus"])\
                                .apply(lambda rows: 
                                      pd.Series(aggregate_results_scweat(rows.sort_values("iter")['SC-WEAT'].tolist())) )
#results_scweat_groups = results_scweat_groups.reset_index()
results_scweat_groups.sort_index(level=['weat_test_num'])
```

### SWEAT

```python
results_sweat = pd.read_json("../results/experiment_W2V_several_runs_WEAT_2/WEAT/sweat_test_3.json")
results_sweat.head(3)
```

```python
# explicit attrib and target set names
results_sweat = results_sweat.merge(results_sweat.set_names.apply(lambda sn: pd.Series(sn)), 
                                          left_index=True, right_index=True)
results_sweat = results_sweat.drop(columns=['test_name'])
```

```python
def aggregate_results_sweat(sweat_x, sweat_y): 
    
    
    
    res = {}
    for i, sweat in zip(['X', 'Y'], [sweat_x, sweat_y]):
    
        score = sweat[0]['score']
        ps = np.array([w['p'] for w in sweat])

        if (ps<=0.05).all():
            ps = "**"
        elif (ps<=0.10).all():
            ps = "*"
        else:
            ps = ""
            
        
        res[f'score_{i}'] = score
        res[f'p_{i}'] = ps
    return res
    
```

```python
results_sweat_groups = results_sweat.groupby(['A', 'B', 'X', 'Y', "weat_test_num", "corpus"])\
                                .apply(lambda rows: 
                                      pd.Series(aggregate_results_sweat(rows.sort_values("iter")['SWEAT_X'].tolist(),
                                                                       rows.sort_values("iter")['SWEAT_Y'].tolist())) )
#results_sweat_groups = results_sweat_groups.reset_index()
results_sweat_groups.sort_index(level=['weat_test_num'])
```

### Merge all tables

```python
res_weat = results_weat_groups.sort_index(level=['weat_test_num'])
```

```python
res_weat['effect_size'] = res_weat.apply(lambda row: "{:.2f}{}".format(row.effect_size, row.p), axis=1)
res_weat = res_weat.drop(columns=['p'])
```

```python
res_scweat = results_scweat_groups.sort_index(level=['weat_test_num'])
```

```python
res_scweat['scweat X'] = res_scweat.apply(lambda row: "{:.2f}{}".format(row.score_X, row.p_X), axis=1)
res_scweat['scweat Y'] = res_scweat.apply(lambda row: "{:.2f}{}".format(row.score_Y, row.p_Y), axis=1)
res_scweat = res_scweat.drop(columns=['score_X', 'p_X', 'score_Y', 'p_Y'])
```
