```python
# match the two tables
# match first the authors and then the song titles 

# lower case all names and titles
test_sexism.loc[:, 'artist_name'] = test_sexism.artist_name.apply(lambda name: name.lower())
test_sexism.loc[:, 'song_title'] = test_sexism.song_title.apply(lambda title: title.lower())

song_lyrics_sexism.loc[:, 'artist_name'] = song_lyrics_sexism.artist_name.apply(lambda name: name.lower())
song_lyrics_sexism.loc[:, 'song_title'] = song_lyrics_sexism.song_title.apply(lambda title: title.lower())

# in WASABi, sometimes song titles like beyoncé:party (with artist_name:song_title)
song_lyrics_sexism.loc[:, 'song_title'] = song_lyrics_sexism.song_title.apply(lambda st: 
                                                        st.split(":")[1] if len(re.findall("\S:\w", st))>0 else st )

# in song title, replace & with "and"
song_lyrics_sexism.loc[:, 'song_title'] = song_lyrics_sexism.song_title.apply(lambda st: 
                                                " and ".join(map(lambda s: s.strip(), st.split("&"))) if "&" in st else st )
test_sexism.loc[:, 'song_title'] = test_sexism.song_title.apply(lambda st: 
                                                " and ".join(map(lambda s: s.strip(), st.split("&"))) if "&" in st else st )

artists_test_set = test_sexism.artist_name.unique()
artists_wasabi = song_lyrics_sexism.artist_name.unique()

candidate_same_artists = []
for artist_test in tqdm(artists_test_set, total=artists_test_set.shape[0]):
    for artist_wasabi in artists_wasabi:

        is_same_name = name_normalizer.same_artist(artist_test, artist_wasabi, with_rotation=False, rotate_stopw=True)

        if is_same_name:
            candidate_same_artists.append({'artist_w':artist_wasabi, 'artist_ts':artist_test})

candidate_same_artists = pd.DataFrame(candidate_same_artists)
candidate_same_artists = candidate_same_artists.groupby("artist_ts")\
                            .apply(lambda rows: rows.artist_w.tolist())

print("Number of unique test artists matched: ", candidate_same_artists.shape[0])
print("Total number of test artists: ", artists_test_set.shape[0])
```

#### Match song titles

```python
# now match titles

candidate_matches = []
for artist_ts, artists_w in tqdm(candidate_same_artists.items(), total=candidate_same_artists.shape[0]):
    
    song_titles_ts = test_sexism[test_sexism.artist_name==artist_ts]
    song_titles_w = song_lyrics_sexism[song_lyrics_sexism.artist_name.isin(artists_w)]

    for idx_ts, row_ts in song_titles_ts.iterrows():
        for idx_w, row_w in song_titles_w.iterrows():

            song_title_ts = row_ts.song_title
            song_title_w = row_w.song_title

            is_same_title = name_normalizer.same_title(song_title_ts, song_title_w)

            if is_same_title:
                candidate_matches.append({'song_title_w':song_title_w,
                                          'song_title_ts':song_title_ts,
                                          'artist_name_w':row_w.artist_name,
                                          'artist_name_ts':row_ts.artist_name,
                                          'song_id_w':row_w.song_id,
                                          'song_id_ts':row_ts.song_id})
                
candidate_matches = pd.DataFrame(candidate_matches)
candidate_matches_grouped = candidate_matches.groupby(["song_title_ts", 'artist_name_ts', 'song_id_ts'])\
                                        .apply(lambda rows: pd.Series({'songs_w':rows[['song_title_w', 'artist_name_w', 
                                                                  'song_id_w']].values.tolist()}))\
                                        .reset_index()

print("Number of unique test set songs matched: ", candidate_matches.song_id_ts.nunique())
```

### Classification metrics

#### Discard feminist label

```python
def bootstrap_conf_int(data, metric, n_iter=1000, alpha=0.05, return_bootstraps=False, rng=None):
    '''
    data : pd.DataFrame with 'is_sexist' and 'label_id' columns
    '''

    score = metric(data.label_id, data.is_sexist)

    scores_boot = []
    for _ in range(n_iter):
        data_boot = data.sample(frac=1., replace=True, random_state=rng)
        score_boot = metric(data_boot.label_id,
                            data_boot.is_sexist)
        scores_boot.append(score_boot)
    scores_boot = np.array(scores_boot)

    CI_l = 2*score - np.quantile(scores_boot, q=1-alpha/2)
    CI_u = 2*score - np.quantile(scores_boot, q=alpha/2)
    
    if return_bootstraps:
        return score, CI_l, CI_u, scores_boot
    else:
        return score, CI_l, CI_u
```

```python
dataset_nofeminist = dataset[dataset.label!='feminist']

print("Number of test instances: ", dataset_nofeminist.shape[0])
print("Class distribution: ")
print(dataset_nofeminist.label.value_counts(1))
```

```python
# 95% confidence interval: bootstrap -> median +- (2.5 and 97.5 percentiles)
label2id = {
    'neutral':0,
    'sexist':1,
}

performances = {}
thresholds = np.arange(50, 100, 2.5) / 100

score_names = ['Precision', 'Recall', 'F1-score', 
               'Accuracy', 'Balanced Accuracy (adjusted)',
               'Precision (macro)', 'Recall (macro)', 'F1-score (macro)',
               'Precision (neg class)', 'Recall (neg class)', 'F1-score (neg class)',]
metric_functions = [precision_score, recall_score, f1_score, 
                    accuracy_score, balanced_accuracy_score,
                    precision_score, recall_score, f1_score,
                    precision_score, recall_score, f1_score]

for threshold in thresholds:

    # get predictions
    predictions = dataset_nofeminist[['sexist_lines', 'label']]

    label_pred = predictions.sexist_lines.apply(lambda lines: 
                                            1 if any([True for l in lines if l[0]>=threshold]) else 0 )
    label_true = predictions.label.apply(lambda l: label2id[l])
    
    labels = pd.concat([label_pred, label_true], axis=1)
    labels.columns = ['is_sexist', 'label_id']

    # compute and store metrics
    performances[threshold] = {}

    for metric_func, metric_name in zip(metric_functions, score_names):
        
        scores = []
        if metric_name in ['Precision', 'Recall', 'F1-score']:
            pos_label = 1

            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, pos_label=pos_label, zero_division=0)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)
        
        elif metric_name in ['Precision (neg class)', 'Recall (neg class)', 'F1-score (neg class)']:
            pos_label = 0

            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, pos_label=pos_label, zero_division=0)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        elif metric_name in ['Precision (macro)', 'Recall (macro)', 'Specificity (macro)', 'F1-score (macro)']:
            #pos_label = 1 if metric_name!='Specificity (macro)' else 0
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, zero_division=0, average='macro')
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        elif metric_name=='Balanced Accuracy (adjusted)':
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, adjusted=True)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        else:
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        performances[threshold][metric_name] = score_true
        performances[threshold][metric_name+'_lb'] = score_true - score_lb
        performances[threshold][metric_name+'_ub'] = score_true_ub - score_true
        
  
```

```python
plt.figure(figsize=(15,20))

for n, score_name in enumerate(score_names, 1):

    plt.subplot(4,3,n)
    #plt.plot(thresholds, [performances[thr][score_name] for thr in thresholds])
    scores = [performances[thr][score_name] for thr in thresholds]
    max_at = np.argmax(scores)
    best_threshold = thresholds[max_at]

    plt.errorbar(thresholds, 
                 scores,
                 yerr=np.vstack([[performances[thr][score_name+'_lb'] for thr in thresholds],
                                 [performances[thr][score_name+'_ub'] for thr in thresholds]]),
                 capsize=5,
                 fmt='o',
                 markersize=5,
                 ls='dashed')
    
    if score_name=='F1-score':
        frac_pos_class = dataset_nofeminist.label.value_counts(1)['sexist']
        baseline = 2 * frac_pos_class / (1 + frac_pos_class)
        plt.hlines(baseline,
                  thresholds[0], thresholds[-1], color='k', linestyle='dashed')

    plt.xlabel('Classification threshold', fontsize=14)
    plt.ylabel(score_name, fontsize=14)
    plt.grid()
    plt.ylim(-0.02, 1.02)

    print(f'{score_name}: {round(max(scores), 4)} at threshold {best_threshold}')

plt.tight_layout()
```

```python
# 95% confidence interval: bootstrap -> median +- (2.5 and 97.5 percentiles)
label2id = {
    'neutral':0,
    'sexist':1,
    'feminist':1
}

performances_femin_sexs = {}
thresholds = np.arange(50, 100, 2.5) / 100

score_names = ['Precision', 'Recall', 'F1-score', 
               'Accuracy', 'Balanced Accuracy (adjusted)',
               'Precision (macro)', 'Recall (macro)', 'F1-score (macro)',
               'Precision (neg class)', 'Recall (neg class)', 'F1-score (neg class)',]
metric_functions = [precision_score, recall_score, f1_score, 
                    accuracy_score, balanced_accuracy_score,
                    precision_score, recall_score, f1_score,
                    precision_score, recall_score, f1_score]

for threshold in thresholds:

    # get predictions
    predictions = dataset[['sexist_lines', 'label']]

    label_pred = predictions.sexist_lines.apply(lambda lines: 
                                            1 if any([True for l in lines if l[0]>=threshold]) else 0 )
    label_true = predictions.label.apply(lambda l: label2id[l])
    
    labels = pd.concat([label_pred, label_true], axis=1)
    labels.columns = ['is_sexist', 'label_id']

    # compute and store metrics
    performances_femin_sexs[threshold] = {}

    for metric_func, metric_name in zip(metric_functions, score_names):
        
        scores = []
        if metric_name in ['Precision', 'Recall', 'F1-score']:
            pos_label = 1

            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, pos_label=pos_label, zero_division=0)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)
        
        elif metric_name in ['Precision (neg class)', 'Recall (neg class)', 'F1-score (neg class)']:
            pos_label = 0

            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, pos_label=pos_label, zero_division=0)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        elif metric_name in ['Precision (macro)', 'Recall (macro)', 'Specificity (macro)', 'F1-score (macro)']:
            #pos_label = 1 if metric_name!='Specificity (macro)' else 0
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, zero_division=0, average='macro')
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        elif metric_name=='Balanced Accuracy (adjusted)':
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred, adjusted=True)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        else:
            metric_wrap = lambda y_true, y_pred: metric_func(y_true, y_pred)
            score_true, score_lb, score_true_ub = bootstrap_conf_int(labels, metric_wrap, n_iter=1000, alpha=0.05)

        performances_femin_sexs[threshold][metric_name] = score_true
        performances_femin_sexs[threshold][metric_name+'_lb'] = score_true - score_lb
        performances_femin_sexs[threshold][metric_name+'_ub'] = score_true_ub - score_true
        
  
```

```python
plt.figure(figsize=(15,20))

for n, score_name in enumerate(score_names, 1):

    plt.subplot(4,3,n)
    scores = [performances_femin_sexs[thr][score_name] for thr in thresholds]
    max_at = np.argmax(scores)
    best_threshold = thresholds[max_at]

    plt.errorbar(thresholds, 
                 [performances_femin_sexs[thr][score_name] for thr in thresholds],
                 yerr=np.vstack([[performances_femin_sexs[thr][score_name+'_lb'] for thr in thresholds],
                                 [performances_femin_sexs[thr][score_name+'_ub'] for thr in thresholds]]),
                 capsize=5,
                 ls='dashed')
    
    if score_name=='F1-score':
        frac_pos_class = dataset.label.value_counts(1).loc[['sexist', 'feminist']].sum()
        baseline = 2 * frac_pos_class / (1 + frac_pos_class)
        plt.hlines(baseline,
                  thresholds[0], thresholds[-1], color='k', linestyle='dashed')

    plt.xlabel('Classification threshold', fontsize=14)
    plt.ylabel(score_name, fontsize=14)
    plt.grid()
    plt.ylim(-0.02, 1.02)

    print(f'{score_name}: {round(max(scores), 4)} at threshold {best_threshold}')

plt.tight_layout()
```
